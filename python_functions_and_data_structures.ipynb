## Goal 1: scaling functions

Write a function that prints all prime numbers from 1 to up to the input value of the function.

```python
# tafera 1a

def primos (x):
  lista = [1]
  for i in range(2, x):
    primo = True
    for j in range (2, i):
      if (i%j == 0):
        primo = False
    if primo:
        lista.append(i)
  return lista

primos(10)
```

Given a vector obtained from the function developed in goal 1, this function should return the following statistics in a `pd.Series` object with corresponding index labels: number of elements, sum, mean, median, variance, standard deviation and any any other stats you find useful.

## Goal 6:

1. Using the variable `Age` from the BES dataset, calculate the age of each respondent rounded to the nearest multiple of 5. Try writing this using a defined function and with a `lambda` function.
2. Recode column `y09` to 0 and 1.
3. Write a function that takes the lower bound of the income limits given in column `y01` and returns it as an integer.


```python
# upload dataset

df = pd.read_feather("data/03_py4ps_bes_data_subset.feather")
```

```python
# solution 1

idades = df['Age']
prio = np.array([1, -1, 2, -2, 3, -3, 4, -4]) # fiz uma ordem de prioridade: para chegar ao múltiplo de 5 mais próximo, tenho que somar com os seguintes números nessa ordem
prio1 = pd.Series(prio)
cond = prio1.dropna().astype(int)
idades5 = []


for x in idades:
  if math.isnan(x) == True:
    idades5.append(np.nan)
  if x%5 == 0:
    idades5.append(x)
  else:
    for j in cond:
      if (j+x)%5 == 0:
        idades5.append(j+x)
        break


print (idades5)
print (type (idades5), len(idades5))
len(idades5) == len(idades)
```
