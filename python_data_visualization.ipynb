# Data visualization with Python

I'm using python programming to describe the electoral preferences of britsh people.

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

link = 'data/py4ps_bes_data_subset.feather'
db = pd.read_feather(link)

dictforbes = {
    'Age':'age',
    'y01':'income_band',
    'y03':'home_situation',
    'y06':'religion',
    'y09':'sex',
    'y17':'job'
}

db = db.rename(dictforbes, axis=1)

a = list(dictforbes.values())

dbfacet = db[a]
```

## Goal 1: a simple electoral distribution according to age.

```python
pal = sns.color_palette("tab10")
f, plot = plt.subplots(1, 1, figsize=(8, 16))
sns.swarmplot(data=db, y=db['a02'], x=db['age'],
              palette=pal)
plot.set(ylabel='Best Party', xlabel='Age')
plt.show()
```

## Goal 2: Panelling with For Loops


```python
f, ax = plt.subplots(1, 11, figsize=(20,8), sharey=True)

for i, job in enumerate(db['job'].unique()):
    temp = db[db['job']==job]
    sns.swarmplot(x=temp['sex'], y=temp['age'], ax=ax[i], s=4)
    ax[i].yaxis.set_label_text('Age', visible=False)
    ax[i].xaxis.set_label_text(job, rotation=90)
```
